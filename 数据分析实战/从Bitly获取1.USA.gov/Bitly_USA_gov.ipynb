```python
def get_count(sequence):
    counts={}
    for x in sequence:
        if x in counts:
            counts[x]+=1
        else:
            counts[x]=1
    return counts
```

```python
from collections import defaultdict
def get_counts(sequence):
    counts = defaultdict(int)
    for x in sequence:
        counts[x]+=1
    return counts
```

```python
counts = get_count(time_zones)
```

```python
cframe = frame[frame.a.notnull()]
```

```python
import numpy as np
cframe['os'] = np.where(cframe['a'].str.contains('Windows'),'Windows','Not Windows')
```

```python
cframe['os'][:5]
```

```python
by_tz_os = cframe.groupby(['tz','os'])
```

```python
agg_count = by_tz_os.size().unstack().fillna(0)
```

```python
agg_count[:10]
```

```python
indexer = agg_count.sum(1).argsort()
```

```python
indexer[:10]
```

```python
counts_subset = agg_count.take(indexer[-10:])
```

```python
counts_subset
```

```python
agg_count.sum(1).nlargest(10)
```

```python
counts_subset = counts_subset.stack()
```

```python
counts_subset.name = 'total'
```

```python
counts_subset = counts_subset.reset_index()
```

```python
counts_subset[:10]
```

```python
sns.barplot(x = 'total',y = 'tz',hue = 'os',data = counts_subset)
```

```python
def norm_total(group):
    group['normed_total'] = group.total / group.total.sum()
    return group
```

```python
results = counts_subset.groupby('tz').apply(norm_total)
```

```python
sns.barplot(x = 'normed_total',y = 'tz',hue = 'os',data=results)
```
